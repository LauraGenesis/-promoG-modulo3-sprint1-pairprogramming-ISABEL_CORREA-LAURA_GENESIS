# Ejercicios de limpieza

Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

No tengan espacios.

Estn en minscula.

No tengan parntesis, es decir, quitar "(%)", "(Km2)".

Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

Unidas por _ (sin espacios)

```python
import re
```

```python
# TERMINAR 

columnas = df_total.columns

columnas_nuevas = []

for columna in columnas:
    if '\n' in columna:
        columnas_nuevas.append(columna.replace('\n', '').lower().strip())
    
    elif ':' in columna:
        columnas_nuevas.append(columna.replace(':', '').lower().strip())

    elif '(%)' or '(Km2)' in columna:
        columnas_nuevas.append(columna.replace(r'\(%\)|\(Km2\)', '').lower().strip())
    
    else:
        columnas_nuevas.append(columna.lower().strip())


columnas_nuevas

```

```python
columnas = df_total.columns
patron = re.compile(r"\((.*?)\)")

columnas_nuevas = []

for columna in columnas:
    if '\n' in columna:
        columnas_nuevas.append(columna.replace('\n', '').lower().strip())
    
    elif ':' in columna:
        columnas_nuevas.append(columna.replace(':', '').lower().strip())

    elif re.search(patron, columna):
        re.sub(patron, '', columna.lower().strip())
    else:
        columnas_nuevas.append(columna.lower().strip())


len(columnas_nuevas)
```

Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
# REHACER CUANDO TENGAMOS COLUMNAS

columnas_con_porcentaje = ['Unemployment rate', 'total_tax_rate', 'tax_revenue', 'population_labor_force_participation',
                           'out_of_pocket_health_expenditure', 'gross_tertiary_education_enrollment',
                           'gross_primary_education_enrollment', 'forested_area', 'cpi_change', 'agricultural_land']

for nombre in columnas_con_porcentaje:
    df_total[nombre] = df_total[nombre].str.replace('%', '')

df_total
```

Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columnas_dolar = ['Gasoline Price', 'GDP', 'Minimum wage']

for nombre in columnas_dolar:
    df_total[nombre] = df_total[nombre].str.replace('$', '')

df_total.head()
```
