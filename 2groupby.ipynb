1. Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
lista1 = ['armed forces size', 'co2 emissions']
lista2 = ['land area', 'gdp', 'population', 'urban population']
```

```python
def to_float1(valor):
    patron = re.compile(r'^\d{1,3}(,\d{3}){2,}(\.\d+)?$')
    if type(valor) == str:
        if re.search(patron, valor):
            return valor.replace(',','', 1).replace(',', '.')
        
        else:
            return float(valor.replace(',','.'))
    else:
        return np.nan
```

```python
for columna in lista1:
    df[columna] = df[columna].apply(to_float1)

df.head()
```

```python
def to_float2(valor):
    if type(valor) == str:
            return float(valor.replace(',',''))
    else:
        return np.nan
```

```python
for columna in lista2:
    df[columna] = df[columna].apply(to_float2)

df.head()
```
