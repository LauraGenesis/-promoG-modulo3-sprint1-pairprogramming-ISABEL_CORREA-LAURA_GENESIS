```python
df = df1.merge(df2, left_on='Country', right_on='country')
df.head()
```

```python
#1.Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.
df.drop(['country', 'Largest city'], axis=1, inplace=True)
df.shape
```

Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

- No tengan espacios.

- Estén en minúscula.

- No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

- Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

- Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
df.columns
```

```python
nuevas_columnas = []
columnas = []
patron = re.compile(r'\([^)]*\)')

for columna in df.columns:
      
    if re.search(patron, columna):
        columna = columna.split('(')
        nuevas_columnas.append(columna[0].strip().lower())
    else:
        nuevas_columnas.append(columna.lower().strip())


for columna in nuevas_columnas:
    if ':' in columna:
        columnas.append(columna.replace(':', ''))
    elif '_' or '-' in columna:
        columnas.append(columna.replace('_', ' ').replace('-', ' '))
    else:
        columnas.append(columna)

```

```python
columnas
```

```python
dicc = {}

for og, nv in zip(df.columns, columnas):
    dicc[og] = nv

dicc
```

```python
df.rename(dicc, axis=1, inplace=True)
```

```python
df.head()
```

La columnas `coordinates` tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la `longitud` y otra con la `latitud`. Una vez hecho, elimina la columna de coordinates.

```python
df['coordinates'] = df['coordinates'].str.replace("(", "").str.replace(")", "").str.replace("'", "")
df['coordinates'] 
```

```python
df[['latitude', 'longitude']] = df['coordinates'].str.split(' , ', expand=True).get([0,1])
```

```python
df.drop('coordinates', axis=1, inplace=True)
```

Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
lista_percent = ['unemployment rate', 'total tax rate', 'tax revenue', 'population labor force participation', 'out of pocket health expenditure', 'gross tertiary education enrollment', 'gross primary education enrollment', 'forested area', 'cpi change']

for columna in lista_percent:
    df[columna] = df[columna].str.replace('%', '')
```

```python
df.sample()
```

Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
lista_dolar = ['gasoline price', 'gdp', 'minimum wage']

for columna in lista_dolar:
    df[columna] = df[columna].str.replace('$','')
```
